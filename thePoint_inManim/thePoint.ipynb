```python
from manim import *
```

```python
%%manim -qm -v WARNING Punto
config.media_width = "75%"
class Punto(Scene):
    def construct(self):
        #creamos el objeto punto de radio 0.5 color rojo
        punto = Dot(radius = 0.5, color = RED)
        #colocamos el objeto punto en la escena
        self.add(punto)
```

```python
%%manim -qm -v WARNING Punto2
config.media_width = "75%"
class Punto2(Scene):
    def construct(self):
        #creamos el objeto punto de radio 1.0 y color verde 
        punto = Dot(radius = 1.0, color = GREEN)
        #animación el objeto en la escena
        self.play(Create(punto))
        self.wait(2)
```
